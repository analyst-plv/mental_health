# Факторы, влияющие на психическое здоровье

О наборе данных
Подробные сведения о психическом здоровье: разнообразный набор данных о 1000 людях из разных профессий, стран и с разным образом жизни

Этот набор данных представляет собой обширную коллекцию анонимных данных о психическом здоровье 1000 человек, представляющих широкий возрастной диапазон, пол, род занятий и страны. Он призван пролить свет на различные факторы, влияющие на психическое здоровье, и предоставить ценную информацию об уровне стресса, режиме сна, балансе между работой и личной жизнью и физической активности.

Ключевые особенности:
Демографические данные: набор данных включает информацию о людях из разных стран, таких как США, Индия, Великобритания, Канада и Австралия. В каждой записи содержится ключевая демографическая информация, такая как возраст, пол и род занятий (например, информационные технологии, здравоохранение, образование, инженерия).

Проблемы с психическим здоровьем: набор данных содержит информацию о том, сообщали ли люди о каких-либо проблемах с психическим здоровьем (да/нет), а также о степени тяжести этих проблем, которая может быть низкой, средней или высокой.

История консультаций: для людей с психическими расстройствами в наборе данных указано, обращались ли они за консультацией к специалисту в области психического здоровья.

Уровень стресса: уровень стресса у каждого человека классифицируется как низкий, средний или высокий, что позволяет понять, как различные факторы, такие как продолжительность рабочего дня или сна, могут влиять на психическое благополучие.

Факторы образа жизни: набор данных включает информацию о продолжительности сна, количестве рабочих часов в неделю и еженедельном количестве часов физической активности, что позволяет получить подробное представление о том, как факторы образа жизни влияют на психическое здоровье.

Этот набор данных можно использовать для исследований, анализа или создания моделей машинного обучения для прогнозирования тенденций в области психического здоровья, выявления взаимосвязей между балансом между работой и личной жизнью и психическим благополучием, а также для изучения влияния стресса и физической активности на психическое здоровье.

## Исследовательский анализ

### Средний возраст

```python
print('Средний возраст человека в данных:',int(data['age'].mean().round()))
```

### Распределение возраста людей, имеющих и не имеющих проблемы с психическим здоровьем


```python
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='age', hue='mental_health_condition', stat="density", common_norm=False, kde=True, bins=50)
plt.grid(True)
plt.title('Распределение возраста')
plt.xlabel('возраст')
plt.ylabel('частота')
plt.legend(['имеющие проблемы','не имеющие']) 
plt.show()
```

На графике видно, что люди, которые сообщали о каких-либо проблемах с психическим здоровьем - это чаще всего люди среднего возраста от 33 до 45 лет, а также небольшое превышение частоты обращений - людей старшего возраста, от 57 лет.

### Распределение продолжительности сна у людей, имеющих и не имеющих проблемы с психическим здоровьем

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='sleep_hours', hue='mental_health_condition', stat="density", common_norm=False, kde=True, bins=30)
plt.grid(True)
plt.title('Распределение продолжительности сна')
plt.xlabel('продолжительность сна')
plt.ylabel('частота')
plt.legend(['имеющие проблемы','не имеющие']) 
plt.show()
```

На графике видно, что люди, которые спят меньше 6 часов в сутки, больше подвержены психическим расстройствам. 

### Распределение продолжительности рабочей недели в часах у людей, имеющих и не имеющих проблемы с психическим здоровьем

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='work_hours', hue='mental_health_condition', stat="density", common_norm=False, kde=True, bins=30)
plt.grid(True)
plt.title('Распределение продолжительности рабочей недели в часах')
plt.xlabel('продолжительность рабочей недели в часах')
plt.ylabel('частота')
plt.legend(['имеющие проблемы','не имеющие']) 
plt.show()
```

Люди, продолжительность рабочей недели которых превышает 45 часов, чаще имеют проблемы с психическим здоровьем. За исключением продолжительности работы от 58 до 68 часов. Интересно, каким видом деятельности они занимаются?

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='physical_activity_hours', hue='mental_health_condition', stat="density", common_norm=False, kde=True)
plt.grid(True)
plt.title('Распределение продолжительности физической активности в часах')
plt.xlabel('продолжительность физической активности в часах')
plt.ylabel('частота')
plt.legend(['имеющие проблемы','не имеющие']) 
plt.show()
```

На графике видно, что люди, ведущие малоактивный образ жизни (до 1,5 часов активных часов) и люди, имеющие физическую активность более 6,5 часов в день -чаще имеют проблемы с психическим здоровьем.

Построим столбчатые диаграммы

```python
average_percentage = round(100*data['mental_health_condition'].count(),2)
average_percentage
```

```python
parameters = ['gender','occupation','country','mental_health_condition','severity','consultation_history','stress_level']

for i in range(len(parameters)):
    data= pd.pivot_table(data, index=parameters[i], columns='mental_health_condition', values='user_id', aggfunc='count').reset_index()
    data.columns=[i, 'loyal', 'churn']
    data['percent_churn']=round(100*data['mental_health_condition'].count()/(data['mental_health_condition'].count()+data['loyal'].count()),2)
    print(data)
    print('средний процент оттока клиентов по банку',average_percentage)
    ax = sns.barplot(x=i, y='percent_churn', data=data) 
    plt.title(f'Доля отточных клиентов по {parameters[i]}')
    plt.ylabel('Процент оттока')
    plt.xlabel(f'параметр {parameters[i]}')
    plt.axhline(average_percentage, color='r', linestyle='--')
    plt.show()
```
